```python
import pandas as pd

df_x=pd.read_csv("/home/abhay/Desktop/Files/machine_learning_challenge/beginner/data/trainnew.csv")
df_y=pd.read_csv("/home/abhay/Desktop/Files/machine_learning_challenge/beginner/data/train_y.csv")
test_x=pd.read_csv("/home/abhay/Desktop/Files/machine_learning_challenge/beginner/data/testnew.csv")
```

```python
import numpy as np
from numpy.linalg import inv
print(np.shape(df_x))
print(np.shape(df_y))
print(np.shape(test_x))
```

```python
x_trans=df_x.transpose()
a=np.matmul(x_trans,df_x)
b=inv(a)
c=np.matmul(b,x_trans)
d=np.matmul(c,df_y)
print(np.shape(d))
```

```python
y_test=np.zeros((255,1))
y_test=np.matrix(y_test)
print(np.shape(y_test))
x=np.matrix(test_x)
i=0
for i in range(0,255):
    x_test=x[i,]
    x_test=np.matrix(x_test)
    predict=np.matmul(x_test,d)
    y=np.abs(predict)
    y_test[i]=np.round(y)
#print(y_test)
np.savetxt('text.csv', y_test, delimiter=',')
#np.to_csv("text.csv")
```
